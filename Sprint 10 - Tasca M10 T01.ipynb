```python
from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings('ignore')
import requests
import pandas as pd
```

## Beautiful Soup:

#### a) http://quotes.toscrape.com

```python
url_a = "http://quotes.toscrape.com"
```

```python
r = requests.get(url_a)
```

```python
r.status_code
```

```python
soup = bs(r.content)
```

```python
print(soup.prettify()) 
```

```python
quote_elements = soup.find_all("div", class_="quote")
quote_elements
```

```python
df = pd.DataFrame()
for quote_element in quote_elements:
    quote = quote_element.find("span", class_="text")
    author = quote_element.find("small", class_="author")
    df = df.append({"Frase" : quote.text.strip(),
                            "Autor" : author.text.strip()},
                        ignore_index=True)
```

```python
df
```

Recogemos en un dataframe las frases y el autor de la primera página de la web a la que hemos realizado web scraping con Beautiful Soup.
